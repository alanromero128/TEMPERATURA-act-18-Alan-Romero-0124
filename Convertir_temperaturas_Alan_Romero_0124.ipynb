Ejemplos Resueltos Paso a PasoEjemplo A: Convertir 25°C a FahrenheitEste es un valor común de "temperatura ambiente".Sustituir en la fórmula: $°F = (25 \times \frac{9}{5}) + 32$Multiplicar: $25 \div 5 = 5$, y $5 \times 9 = 45$.Sumar 32: $45 + 32 = 77$.Resultado: 25°C equivalen a 77°F.Ejemplo B: Convertir 98.6°F a CelsiusEsta es la temperatura promedio del cuerpo humano.Sustituir en la fórmula: $°C = \frac{(98.6 - 32) \times 5}{9}$Restar 32: $98.6 - 32 = 66.6$.Multiplicar por 5/9: $66.6 \times 5 = 333$. Luego $333 \div 9 = 37$.Resultado: 98.6°F equivalen a 37°C.3.
